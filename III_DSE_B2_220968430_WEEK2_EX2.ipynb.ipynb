Q8) Visualize the house price based on build up area of top five locations where the large number of houses available (Visualize by independent charts).

```python
houses_per_area = df['location'].value_counts()

top_five_areas = houses_per_area.nlargest(5).index

top_five_df = df[df['location'].isin(top_five_areas)]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=top_five_df, x='area_type', y='price', hue='location')
plt.title("House Price vs. Build-up Area in Top Five Locations")
plt.xlabel("Area Type")
plt.ylabel("House Price")
plt.legend()
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```
