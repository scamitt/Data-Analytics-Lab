Q23)

```python

matches_played = df['team1'].append(df['team2']).value_counts().reset_index()
matches_played.columns = ['Team', 'Total Matches Played']
matches_won = df['winner'].value_counts().reset_index()
matches_won.columns = ['Team', 'Total Matches Won']
team_stats = pd.merge(matches_played, matches_won, on='Team', how='left')
team_stats['Total Matches Won'].fillna(0, inplace=True)
team_stats['Win Rate'] = (team_stats['Total Matches Won'] / team_stats['Total Matches Played']) * 100
team_stats = team_stats.sort_values(by='Win Rate', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(team_stats['Team'], team_stats['Total Matches Played'], label='Total Matches Played', alpha=0.7)
plt.bar(team_stats['Team'], team_stats['Total Matches Won'], label='Total Matches Won', alpha=0.7)
plt.ylabel('Count')
plt.title('Total Played Matches vs Winning Matches vs Win Rate')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
