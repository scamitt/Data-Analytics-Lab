Q2) For each of the numeric attributes (proteins up to vitamins), identify and replace all missing data (indicated with -1) with the arithmetic mean of the attribute.

```python
values = ['protein','fat','sodium','fiber','carbo','sugars','potass','vitamins']
for key in df:
    if key in values:
        df[key]=df[key].replace(-1,df[key].mean())
 
```

Q4) For each of the numeric attributes (proteins up to vitamins), identify and replace all noisy data with the median of attribute. 

```python
for key in df:
    if key in values:
        q1=np.quantile(df[key],0.25)
        q3=np.quantile(df[key],0.75)
        iqr=q3-q1
        lower = q1-1.5*iqr
        upper=q3+1.5*iqr
        for i in df[key]:
            if i>upper or i<lower:
                df[key]=df[key].replace(i,df[key].median())
```
